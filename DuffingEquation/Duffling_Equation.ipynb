```python
from pylab import *
from scipy.integrate import *
import matplotlib.pyplot 

#parameters
alpha = -1
beta = 1
gamma = 0.71
delta = 1.00005
w = 1

#range of values
t = linspace(0, 40 * pi, 10000)

#initial condition
x0 = [0, 0]

def f(y, t):
    x, v = y
    return [v, -delta * v - beta * x ** 3 - alpha * x + gamma * cos(w * t)]


[x, v] = odeint(f, x0, t).T

#drawing
plot(t, x, color="green", lw=2)
title(" Graph  x(t)")
xlabel("t")
ylabel("x")
xlim(0, 40 * pi)
grid()

rc("figure", figsize=(6.25, 6))
figure()
plot(x, v, color="green", lw=2)
title("Phase plane")
xlabel("x")
ylabel("v")
xlim(-1.5, 1.5)
ylim(-1.5, 1.5)
grid()


plt.show()
```
